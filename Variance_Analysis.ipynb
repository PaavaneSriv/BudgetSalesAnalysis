# **Data cleaning**



## `Dropping irrevalent columns`

- Not all the columns are important to us in the given dataframe, and hence we would drop the columns that are irrevalent to us.


```python
sales = sales[['ProductKey',
 'OrderDate',
 'ShipDate',
 'CustomerKey',
 'SalesTerritoryKey',
 'SalesOrderNumber',
 'SalesOrderLineNumber',
 'OrderQuantity',
 'UnitPrice',
 'TotalProductCost',
 'SalesAmount',
 'TaxAmt',
 'Total sales revenue',
 'Gross Profit ']]
```

## `Renaming Columns`




```python
sales = sales.rename(columns={'TaxAmt':"TaxAmount",'Total sales revenue':"TotalSalesRevenue",'Gross Profit':"GrossProfit"})
```

## `Changing the datatype`

- After removing the irrelavant columns, we need to ensure the data types for the columns as it should be. <br>
Checking the datatypes of the revised columns.

- changing the `dtype` of the columns wherever neccesary.

```python
sales['OrderDate'] = pd.to_datetime(sales['OrderDate'],format='mixed')
```

```python
sales['ShipDate'] = pd.to_datetime(sales['ShipDate'],format='mixed')
```
