# **Import the libraries and the Data Set**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import chardet
plt.style.use("seaborn-v0_8-whitegrid")
import warnings
warnings.filterwarnings('ignore')
```

```python
territory= pd.read_csv("/content/Territory BSA.csv",dtype={'SalesTerritoryKey':str})



with open('/content/customer BSA.csv', 'rb') as f:
    result = chardet.detect(f.read())
customer = pd.read_csv("/content/customer BSA.csv",dtype={'CustomerKey':str},
                              parse_dates=['BirthDate','DateFirstPurchase'],
                              encoding=result['encoding'])

product = pd.read_csv("/content/product BSA.csv",dtype={'ProductKey':str},
                      parse_dates=['StartDate'])

sales = pd.read_csv("/content/sales BSA.csv",
                    dtype={'ProductKey':str,
                          'CustomerKey':str,
                          'PromotionKey':str,
                          'SalesTerritoryKey':str},
                    parse_dates=['OrderDate', 'ShipDate'])
```
